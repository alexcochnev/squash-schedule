```python
response = json.loads(re.sub(r'[)(;]', '', r.get('https://api.sportbs.net/v1_0/model/index/3148').text))
schedule = []
for i in range(14):
    try:
        raw_date = datetime.today() + timedelta(i)
        date = raw_date.strftime('%Y-%m-%d')
        answer = {'date': raw_date.strftime('%d.%m'), 'week': week[raw_date.weekday()]}
        for hour in range(17, 23):
            sec = str(hour * 3600)
            if sec in response['days'][date]['hours'] and len(response['days'][date]['hours'][sec]) == 3:
                answer[str(hour)] = 0
            else:
                answer[str(hour)] = 1
        schedule.append(answer)
    except IndexError:
        pass
print('Ленинка')
schedule
```
